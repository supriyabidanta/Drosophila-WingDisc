```python
import os
import scanpy as sc
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from gprofiler import GProfiler
import squidpy as sq
import scipy
from skimage import filters, measure, morphology
from skimage.color import label2rgb
from skimage.io import imread
from skimage.filters import threshold_otsu
from skimage.morphology import remove_small_objects
```

```python
def load_scRNA_seq_data(sample_name, data_directory):
    matrix_file = os.path.join(data_directory, sample_name, 'matrix.mtx.gz')
    genes_file = os.path.join(data_directory, sample_name, 'genes.tsv.gz')
    barcodes_file = os.path.join(data_directory, sample_name, 'barcodes.tsv.gz')

    adata = sc.read_mtx(matrix_file).T
    adata.var_names = pd.read_csv(genes_file, header=None, sep='\t')[0]
    adata.obs_names = pd.read_csv(barcodes_file, header=None)[0]
    return adata
```

```python
def filter_low_quality_cells(adata, min_genes=200, max_genes=5000, min_counts=500):
    adata.obs['n_counts'] = adata.X.sum(axis=1).A1 if isinstance(adata.X, np.matrix) else np.sum(adata.X, axis=1)
    adata.obs['n_genes_by_counts'] = (adata.X > 0).sum(axis=1).A1 if isinstance(adata.X, np.matrix) else (adata.X > 0).sum(axis=1)

    mito_genes = adata.var_names.str.startswith(('mt-', 'mt:')).to_numpy()
    
    if mito_genes.sum() > 0:
        adata.var['mt'] = mito_genes
        sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
        print(f"Found {mito_genes.sum()} mitochondrial genes for QC.")
    else:
        print("No mitochondrial genes found — skipping mito QC.")
        sc.pp.calculate_qc_metrics(adata, percent_top=None, log1p=False, inplace=True)

    initial_cell_count = adata.n_obs

    sc.pp.filter_cells(adata, min_genes=min_genes)
    sc.pp.filter_cells(adata, min_counts=min_counts)
    sc.pp.filter_cells(adata, max_genes=max_genes)

    final_cell_count = adata.n_obs
    print(f"Filtered out {initial_cell_count - final_cell_count} low-quality cells.")

    return adata
```

```python
def preprocess_data_for_visualization(adata):
    sc.pp.normalize_total(adata, target_sum=1e4)
    sc.pp.log1p(adata)
    sc.pp.highly_variable_genes(adata, flavor='seurat_v3', n_top_genes=3000)
    sc.tl.pca(adata, svd_solver='arpack')
    sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)
    sc.tl.umap(adata)
    sc.tl.leiden(adata, resolution=2.0)
    return adata
```

```python
def merge_annotations(data_directory):
    iec_annotations = pd.read_csv(os.path.join(data_directory, 'iEC_barcodes_tissueAnnotations.csv'))
    ctrl_annotations = pd.read_csv(os.path.join(data_directory, 'Ctrl_barcodes_tissueAnnotations.csv'))

    iec_annotations['Unnamed: 0'] = 'iEC_' + iec_annotations['Unnamed: 0'].astype(str)
    ctrl_annotations['Unnamed: 0'] = 'Ctrl_' + ctrl_annotations['Unnamed: 0'].astype(str)

    iec_annotations['condition'] = 'iEC'
    ctrl_annotations['condition'] = 'Ctrl'

    merged_annotations = pd.concat([iec_annotations, ctrl_annotations])
    merged_annotations = merged_annotations.set_index('Unnamed: 0')

    assert merged_annotations.index.is_unique, "Merged annotation barcodes are not unique!"

    return merged_annotations
```

```python
def analyze_merged_sample(data_directory, sample_name, output_dir, epithelial_region_markers):
    print(f'Running analysis for merged sample: {sample_name}')

    merged_annotations = merge_annotations(data_directory)
    adata = load_scRNA_seq_data(sample_name, data_directory)

    adata.obs['cell_type'] = adata.obs_names.map(merged_annotations['celltype']).fillna('Unknown')
    adata.obs['condition'] = adata.obs_names.map(merged_annotations['condition']).fillna('Unknown')

    adata = filter_low_quality_cells(adata)
    adata = preprocess_data_for_visualization(adata)

    return adata
```

```python
def get_valid_genes(adata, genes):
    """
    Filter input genes to those present in adata.var_names.
    Returns valid genes and missing genes.
    """
    valid = [g for g in genes if g in adata.var_names]
    missing = [g for g in genes if g not in adata.var_names]
    
    if missing:
        print(f"Missing genes (not found in adata.var_names): {missing}")
    
    return valid


def plot_umap_by_genes(adata, genes, output_dir, sample_id, group_size=5):
    """
    Plot UMAPs of gene expressions in groups.
    Splits into multiple plots if too many genes.
    """
    os.makedirs(output_dir, exist_ok=True)
    valid_genes = get_valid_genes(adata, genes)
    
    if not valid_genes:
        print("No valid genes found for UMAP plotting.")
        return

    # Break genes into smaller groups
    for i in range(0, len(valid_genes), group_size):
        gene_batch = valid_genes[i:i + group_size]
        save_name = f"_{sample_id}_umap_markers_batch{i//group_size+1}.png"
        
        print(f"Plotting UMAP for genes: {gene_batch}")
        sc.pl.umap(
            adata, color=gene_batch, cmap='viridis',
            save=save_name, show=False
        )

```

```python
def visualize_data(adata, sample_id, output_dir, marker_genes=None):
    os.makedirs(output_dir, exist_ok=True)
    sc.settings.figdir = output_dir

    # Create subfolders
    umap_dir = os.path.join(output_dir, "UMAPs")
    qc_dir = os.path.join(output_dir, "QC")
    markers_dir = os.path.join(output_dir, "Markers")
    densities_dir = os.path.join(output_dir, "Densities")
    dotplot_dir = os.path.join(output_dir, "Dotplots")
    heatmap_dir = os.path.join(output_dir, "Heatmaps")
    stackedbar_dir = os.path.join(output_dir, "Celltype_Distribution")

    for folder in [umap_dir, qc_dir, markers_dir, densities_dir, dotplot_dir, heatmap_dir, stackedbar_dir]:
        os.makedirs(folder, exist_ok=True)

    sc.settings.figdir = umap_dir

    # --- UMAPs ---
    print(f"UMAP by cell type...")
    sc.pl.umap(adata, color="cell_type", palette="tab20", save=f"_{sample_id}_celltype.png", show=False)

    print(f"UMAP by condition...")
    if "condition" in adata.obs.columns:
        sc.pl.umap(adata, color="condition", palette="Set2", save=f"_{sample_id}_condition.png", show=False)

    print(f"UMAP by Leiden clusters...")
    if "leiden" in adata.obs.columns:
        sc.pl.umap(adata, color="leiden", palette="tab20", save=f"_{sample_id}_leiden.png", show=False)

    if marker_genes is not None:
        print(f"UMAPs for marker genes...")
        plot_umap_by_genes(adata, marker_genes, output_dir=umap_dir, sample_id=sample_id, group_size=4)

    # --- QC Metrics ---
    sc.settings.figdir = qc_dir
    print(f"Violin plot for QC metrics...")
    sc.pl.violin(adata, keys=['n_genes_by_counts', 'n_counts'], groupby='cell_type', save=f"_{sample_id}_qc_violin.png", show=False)

    # --- Marker Analysis ---
    sc.settings.figdir = markers_dir
    print(f"Rank marker genes...")
    sc.tl.rank_genes_groups(adata, groupby="leiden", method="wilcoxon")
    sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False, save=f"_{sample_id}_ranked_markers.png", show=False)

    # --- Dotplot ---
    sc.settings.figdir = dotplot_dir
    if marker_genes is not None:
        valid_genes = get_valid_genes(adata, marker_genes)
        if valid_genes:
            print(f"Dotplot for marker genes...")
            sc.pl.dotplot(adata, valid_genes, groupby='leiden', standard_scale='var', save=f"_{sample_id}_dotplot_markers.png", show=False)
        else:
            print("No valid marker genes for dotplot.")

    # --- Heatmap ---
    sc.settings.figdir = heatmap_dir
    if marker_genes is not None:
        valid_genes = get_valid_genes(adata, marker_genes)
        if valid_genes:
            print(f"Heatmap for marker genes...")
            sc.pl.heatmap(adata, valid_genes, groupby='leiden', swap_axes=True, cmap='viridis', save=f"_{sample_id}_heatmap_markers.png", show=False)
        else:
            print("No valid marker genes for heatmap.")
    # --- Stacked Bar Plot: Celltype distribution by condition ---
    print(f"Stacked barplot: celltype vs condition...")
    sc.settings.figdir = stackedbar_dir
    cross_tab = pd.crosstab(adata.obs['condition'], adata.obs['cell_type'], normalize='index')
    cross_tab.plot(kind='bar', stacked=True, figsize=(10,6))
    plt.title('Cell Type Distribution Across Conditions')
    plt.ylabel('Fraction')
    plt.xlabel('Condition')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig(os.path.join(stackedbar_dir, f"{sample_id}_celltype_distribution.png"))
    plt.close()

    # --- Density Embedding ---
    sc.settings.figdir = densities_dir
    print(f"Embedding density by cell type...")
    try:
        sc.tl.embedding_density(adata, groupby='cell_type')
        sc.pl.embedding_density(adata, basis='umap', groupby='cell_type', save=f"_{sample_id}_embedding_density.png", show=False)
    except Exception as e:
        print(f"Could not plot embedding density: {e}")

    # --- Mitochondrial gene checks ---
    print(f"Checking for mitochondrial genes...")
    mito_genes = [gene for gene in adata.var_names if gene.startswith('mt-') or gene.startswith('mt-col')]

    if mito_genes:
        print(f"UMAPs for mitochondrial genes...")
        sc.pl.umap(adata, color=mito_genes, cmap='inferno', save=f"_{sample_id}_mito_genes.png", show=False)
    else:
        print("No mitochondrial genes found for plotting.")

    print("All visualizations complete.")

```

```python
def run_dge_per_celltype(adata, groupby='condition', label='cell_type', output_dir='../output/CMG_merged/dge_results/'):

    os.makedirs(output_dir, exist_ok=True)

    unique_celltypes = adata.obs[label].unique()
    all_results = {}

    for ct in unique_celltypes:
        print(f"Running DGE for cell type: {ct}")
        subset = adata[adata.obs[label] == ct]

        if subset.obs[groupby].nunique() < 2:
            print(f"Skipping {ct} — not enough conditions.")
            continue

        sc.tl.rank_genes_groups(subset, groupby=groupby, method='wilcoxon', use_raw=False)
        df = sc.get.rank_genes_groups_df(subset, group=None)  # all groups
        df.to_csv(os.path.join(output_dir, f"DGE_{ct}.csv"), index=False)
        all_results[ct] = df

    print(f"DGE results saved in: {output_dir}")
    return all_results
```

```python
def plot_top_dge_genes(adata, cell_type, groupby='condition', n_genes=10):
    subset = adata[adata.obs['cell_type'] == cell_type]
    sc.tl.rank_genes_groups(subset, groupby=groupby, method='wilcoxon', use_raw=False)
    sc.pl.rank_genes_groups(subset, n_genes=n_genes, sharey=False)
```

## Spatial Mapping

```python
matrix = scipy.io.mmread("../data/Hunter_fruitfly/merged_output//matrix.mtx.gz")
matrix = scipy.sparse.csr_matrix(matrix)

# Barcodes and genes
barcodes = pd.read_csv("../data/Hunter_fruitfly/merged_output/barcodes.tsv.gz", header=None)[0].tolist()
genes = pd.read_csv("../data/Hunter_fruitfly/merged_output/genes.tsv.gz", header=None)[0].tolist()

# Create AnnData
adata = sc.AnnData(X=matrix.T)
adata.obs_names = barcodes
adata.var_names = genes

# Load the tissue image
tissue_img = imread('../data/Hunter_fruitfly/Image/MAXProjection_Slices9through15_crop.tif')
```

```python
def assign_spatial_coordinates_from_regions(adata, tissue_image, region_table_path):
    regions = pd.read_csv(region_table_path, sep="\t")
    
    all_coords = []
    labels = []
    
    n_regions = len(regions)
    base_cells_per_region = adata.n_obs // n_regions
    extra_cells = adata.n_obs % n_regions  # Remaining cells to distribute

    for idx, row in regions.iterrows():
        x_min, x_max = map(int, row['X'].split('-'))
        y_min, y_max = map(int, row['Y'].split('-'))

        n_cells = base_cells_per_region + (1 if idx < extra_cells else 0)

        # Random sample inside the region
        xs = np.random.uniform(x_min, x_max, size=n_cells)
        ys = np.random.uniform(y_min, y_max, size=n_cells)
        
        coords = np.vstack([xs, ys]).T
        
        all_coords.append(coords)
        labels += [row['Region']] * n_cells

    # Stack all
    spatial_coords = np.vstack(all_coords)
    labels = np.array(labels)

    assert spatial_coords.shape[0] == adata.n_obs, \
        f"Mismatch: {spatial_coords.shape[0]} coords vs {adata.n_obs} cells"

    adata.obsm['spatial'] = spatial_coords
    adata.obs['region'] = labels
    
    # Save tissue image
    adata.uns["spatial"] = {
        "fake_library": {
            "images": {"hires": tissue_image},
            "scalefactors": {
                "tissue_hires_scalef": 1.0,
                "spot_diameter_fullres": 8.0
            }
        }
    }
    
    return adata


# Apply
adata = assign_spatial_coordinates_from_regions(
    adata,
    tissue_img,
    region_table_path="../data/Hunter_fruitfly/Image/regions_coordinates.tsv"  # <-- update this to your path
)
```

```python
def create_tissue_mask(image, min_area=1000):
    """Segment tissue area automatically from DAPI image."""
    gray_img = np.mean(image, axis=-1) if image.ndim == 3 else image
    thresh = threshold_otsu(gray_img)
    binary = gray_img > thresh
    
    # Remove small background noise
    binary = remove_small_objects(binary, min_size=min_area)
    
    return binary

def mask_points_inside_tissue(adata, tissue_mask):
    """Mask spatial points that fall inside the real tissue."""
    coords = adata.obsm['spatial'].astype(int)
    image_height, image_width = tissue_mask.shape
    valid = np.zeros(coords.shape[0], dtype=bool)
    
    for i, (x, y) in enumerate(coords):
        if 0 <= x < image_width and 0 <= y < image_height:
            valid[i] = tissue_mask[y, x]  # Note: row=y, col=x
    return valid

def plot_gene_spatial_final(adata, gene, tissue_image, cmap="plasma", size=1.2):
    """Final nice tissue-specific gene plot."""
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Normalize tissue
    if tissue_image.max() > 1.0:
        tissue_image = tissue_image / tissue_image.max()
    
    ax.imshow(tissue_image, alpha=0.7, cmap="gray")
    
    # Create tissue mask
    tissue_mask = create_tissue_mask(tissue_image)
    
    # Mask cells inside tissue
    inside_mask = mask_points_inside_tissue(adata, tissue_mask)
    adata_tissue = adata[inside_mask].copy()
    
    if gene in adata_tissue.var_names:
        expr = adata_tissue[:, gene].X.toarray().flatten() if hasattr(adata_tissue[:, gene].X, "toarray") else adata_tissue[:, gene].X.flatten()
        vmax = np.percentile(expr, 99)
        
        sq.pl.spatial_scatter(
            adata_tissue,
            color=gene,
            size=size,
            ax=ax,
            cmap=cmap,
            alpha=1.0,
            frameon=False,
            colorbar=True,
            library_id="fake_library",
            spatial_key="spatial",
            vmin=0,
            vmax=vmax,
        )
        plt.title(f"{gene} expression", fontsize=14)
        plt.axis('off')
        plt.show()
    else:
        print(f"Gene {gene} not found in dataset.")


# Example usage
plot_gene_spatial_final(
    adata, 
    gene="nub", 
    tissue_image=tissue_img
)
```
